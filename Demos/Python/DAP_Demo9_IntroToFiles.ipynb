# Python File Modes and Error Handling

In Python, when working with files, the way you open a file matters a lot. Different modes determine how you can interact with a file. This section explains the file modes and shows how to safely handle errors using exception handling.

---

Handling Errors and Exceptions
Error handling in Python is done using try...except blocks. This prevents your program from crashing and allows you to handle errors gracefully. Here are a few tips:

Specific Exceptions: Catch specific errors (like FileNotFoundError or FileExistsError) to give meaningful error messages.

General Exceptions: You can catch any exception using except Exception as e:. However, using broad exception handling should be done with caution, so you don't hide unexpected errors.
Example: General Error Handling

```python
# Simple examples of handling errors
#Create a function that take 2 numbers a and b, and return a/b
# First without handling errors
# def divide(a, b):
#     return a / b

#After handling the erros
def divide(a, b):
  try:
      result = a / b
      return result
  except ZeroDivisionError:
      print("You divided by 0. It is not a good idea")
      return None
```

```python
divide(10,0)
```

## File Modes Explained

### Read Mode ('r')

- **Usage**: To read data from a file.

- **Behavior**: The file must exist; otherwise, a `FileNotFoundError` is raised.

```python
with open("ex1.txt", "r") as f: # with helps us to close the file once we read it
  content=f.read() #
  print(content)
```

```python
#Someitmes we call a file that does not exist. So, our code will produce an error
try:
    with open("ex1.txt", "r") as file: # using with to open the file
        content = file.read() # get all the data from file to a variable called content
        print("File content:", content) # print the file
except FileNotFoundError:
    print("Error: The file does not exist.")
```

Write Mode ('w')

Usage: To write data to a file.

Behavior: If the file already exists, its content is overwritten. If it does not exist, a new file is created.

```python
DAP_text = "Welcome to DAP Prgram"
with open("dap.txt", "w") as file: # I open the file using w mode. If the file already exist, the new data will be added,
#but the current data will be deleted
    file.write(DAP_text)
print("Data written in write mode.")
```

```python
try:
    with open("ex1.txt", "w") as file:
        file.write("This will overwrite any existing content in example.txt.")
    print("Data written in write mode.")
except Exception as e:
    print("An error occurred:", e)
```

Append Mode ('a')
Usage: To add data to the end of an existing file.
Behavior: If the file doesn't exist, a new file is created.

```python
# Appending to ex1.txt
DAP_text="\nHere is a list of our DAP champions"
with open("dap.txt", "a") as file:
  file.write(DAP_text)
print("Data appended in append mode.")
```

```python
try:
    with open("example.txt", "a") as file:
        file.write("\nThis line is appended to the file.")
    print("Data appended successfully.")
except Exception as e:
    print("An error occurred when appending:", e)
```

Exclusive Creation ('x')
Usage: To create a new file and write data to it.
Behavior: Raises a FileExistsError if the file already exists.

```python
try:
    with open("new_file.txt", "x") as file:
        file.write("This file is created exclusively for this operation.")
    print("File created and written using exclusive mode.")
except FileExistsError:
    print("Error: The file already exists.")
```

Binary Mode ('b') and Text Mode ('t')
Usage: Adding 'b' to the mode opens the file in binary mode (non-text data like images or compiled code), whereas 't' (the default) is used for text files.
Example for binary mode: Opening an image file

```python
try:
    with open("image.png", "rb") as file:
        data = file.read()
        print("Binary data read successfully.")
except FileNotFoundError:
    print("Error: The image file was not found.")
```

```python
try:
    with open("data.txt", "r") as file:
        content = file.read()
        print(content)
except Exception as e:
    print("An unexpected error occurred:", e)
```

Using finally Block
The finally block is optional and is executed whether an exception occurs or not. Its useful for cleaning up resources.

```python
try:
    file = open("data.txt", "r")
    content = file.read()
    print("File content:", content)
except FileNotFoundError:
    print("File not found!")
finally:
    file.close()
    print("The file has been closed.")
```

Saving Data to a JSON File
JSON (JavaScript Object Notation) is a format for storing structured data. It is commonly used for configurations and data exchange.

```python
import json

#Using json.dump help us a lot. There is no need to transform the data in this case
with open("student_info.txt", "w") as file:
    # Convert the list of dictionaries to a JSON string
    json_data = json.dumps(student_info, indent=4)
    # Write the JSON string to the file
    file.write(json_data)
print("Data saved to student_info.txt")
```

C. Saving Data to a CSV File
CSV (Comma-Separated Values) files are used to store tabular data (like spreadsheets).

```python
import csv

# Data to be saved in CSV format
data = [
    ["Name", "Age", "Major"],
    ["Alice", 21, "Physics"],
    ["Bob", 22, "Chemistry"],
    ["Charlie", 20, "Biology"]
]

with open("example.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)
print("Data saved to example.csv")
```
